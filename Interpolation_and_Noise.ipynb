```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

t = np.linspace(0, 10, 10)
y = 500*np.exp(-t)


plt.scatter(t, y)
plt.show()
```

```python
finterp = interp1d(t, y)

finterp_cubic = interp1d(t, y, kind='cubic')

tnew = np.linspace(0, 10, 200)

fig, ax = plt.subplots()
ax.plot(t, y, 'bo')
ax.plot(tnew, finterp(tnew), 'cyan')
ax.plot(tnew, finterp_cubic(tnew), 'red')
ax.set(xlabel = 't', ylabel = 'N')
ax.grid()
plt.show()
```

```python
from scipy.optimize import curve_fit
def y1(t):
    return 500*np.exp(-t)

yvalues = y1(tnew)
rndvalues = np.random.normal(0, 10, len(yvalues))
data_plus_noise = yvalues + rndvalues

plt.plot(tnew, data_plus_noise)
plt.xlabel('t values')
plt.ylabel('Data + Noise')

### curve fitting
def test(t, a):
    return a*np.exp(-t)
params, covar = curve_fit(test, xdata = tnew , ydata = data_plus_noise, p0 = len(tnew), method ='lm')
print(params, covar)
print(np.sqrt(covar))
```
