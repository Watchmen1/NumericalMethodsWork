```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

I) Mass Integration Problem

```python
p_0 = 1
R_s = 3/8

def p(r):
    return p_0 / ((r/R_s) * ((1 + (r/R_s)) ** 2))

def M(r):
    return 4 * np.pi * (r ** 2) * p(r)

x = quad(M, 0, 2)
print(x[0])


Rs = np.linspace(0.01, 2, 1000)
Ms = []

for r in Rs:
    Ms.append(quad(M, 0, r)[0])
    
plt.plot(Rs, Ms)
plt.title('Mass vs r')
plt.xlabel('r')
plt.ylabel('Mass')
plt.show()
```

II) Pendulum Problem

```python
def k(a):
    return np.sin(a / 2)

def integ(phi, a):
    return (2/np.pi) / np.sqrt(1 - (k(a))**2 * np.sin(phi)**2)

a1 = 0.01
small = quad(integ, 0, np.pi/2, args = a1)
print(small[0])
a2 = np.pi/4
larger = quad(integ, 0, np.pi/2, args = a2)
print(larger[0])


r = np.linspace(0, np.pi/4, 1000)
y = []

for a in r:
    y.append(quad(integ, 0, np.pi/2, args=(a))[0])

plt.plot(r,y)
plt.show()
```
