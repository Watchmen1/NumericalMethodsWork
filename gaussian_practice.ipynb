```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def powerlaw(x,a,b):
    return a*np.power(x,b)

xdum = np.linspace(0,1000,100)
ydum = powerlaw(xdum,1, 0.5)
noise = 1.5 * np.random.normal(size=ydum.size)
ydumnoi = ydum+noise

fig, ax = plt.subplots()
ax.plot(xdum, ydum)
ax.plot(xdum, ydumnoi, '.')
ax.set(xlabel='x', ylabel='y')
ax.grid()
```

```python
def gaussian(x, a, b, sig):
    return a*np.exp(-np.power(x - b, 2)/2*np.power(sig, 2))

#dummy dataset
xdum = np.linspace(start=-5, stop=5, num=25)
ydum = gaussian(xdum, 10, 2, 1)

#Add noise from Gaussian Dist. 
noise = 0.5*np.random.normal(size=ydum.size)
ydumpnoi = ydum + noise
```

```python
xdumnew = np.linspace(-5, 5, 1000)
ydumnew = gaussian(xdumnew, 10, 2, 1)

fig, ax = plt.subplots()
ax.plot(xdumnew, ydumnew)
ax.set(xlabel='x', ylabel='y')
ax.plot(xdum, ydumpnoi, '.')
ax.grid()
plt.show()
```
